# data 
scimagoir rank

```python
lists=[]
scimagoircountry=[]  
list2=[]    #school name
response = requests.get('https://www.scimagoir.com/rankings.php?sector=Higher+educ.&ranking=Overall&area=all')
soup=BeautifulSoup(response.text,"html.parser")
x=soup.find_all("td")
count=0

for ele in x: 
  lists.append(ele.text)
for count in range(2,len(lists)):
  if (count-2)%6==0:
    list2.append(lists[count])
  elif  (count-2)%6==1:
    scimagoircountry.append(lists[count])

for c in range (0,len(list2)):
  if list2[c].find("*")>0:
    list2[c]=list2[c].replace("*","")
```

# another csv file,using upload and store into list

```python
import csv
RUR=[]

RURcountry=[]
with open('aaa.csv', newline='') as csvfile:
  reader = csv.DictReader(csvfile)
  for row in reader:
    RUR.append(row['University'])
    RURcountry.append(row['Country'])

# print(RUR)
```

# Normalize data

```python
def normalizeData(arr):
  listnormal=[]
  count=0
  for ele in arr:
    if ele.find("(")>0:
      ele=ele[:ele.index("(")]
    ele=ele.strip()
    listnormal.append(ele)
    count+=1
  return listnormal

RUR=normalizeData(RUR)
list2=normalizeData(list2)
name=normalizeData(name)    #main

print(len(RUR))
print(len(ARWU))
print(len(QS))
print(len(THE))
print(len(ARTU))
print(len(name))
print(len(list2))
print(len(country))

# print (type(RUR[0]))
# print(type(list2[0]))
# print(type(name[0]))

print(RUR)
print(list2)
print(name)

print(RUR[5]==name[2])
```

# Merge Data


```python
RURout=[""]*len(name)
check=0
count=1
for ele in RUR:
  for matchnumber in range(0,len(name)):
    if ele==name[matchnumber]:
      RURout[matchnumber]=count
      check+=1
  if check==0:  
    name.append(ele)
    RURout.append(count)
    ARWU.append("")
    QS.append("")
    THE.append("")
    ARTU.append("")
    country.append(RURcountry[count-1])
  check=0
  count+=1    
# print(len(RURout))
# print(len(ARWU))
# print(len(QS))
# print(len(THE))
# print(len(ARTU))
# print(len(name))
# print(len(scimagoirOut))
# print(len(country))

scimagoirOut=[""]*len(name)
print(len(scimagoirOut))
count=1
for ele in list2:
  for matchnumber in range(0,len(name)):
    if ele==name[matchnumber]:
      scimagoirOut[matchnumber]=count
      check+=1
  if check==0:  
    name.append(ele)
    scimagoirOut.append(count)
    RURout.append("")
    ARWU.append("")
    QS.append("")
    THE.append("")
    ARTU.append("")
    country.append(scimagoircountry[count-1])
  check=0
  count+=1    


```
