```python
import csv
import requests
from bs4 import BeautifulSoup
import re
import sys 
import time
ranking=1
schoolList=[]
website="https://www.niche.com/colleges/search/best-colleges/?page="

for page in range(1,9):
  webPage=website+str(page)
  time.sleep(3)
  # print(webPage)
  response = requests.get(webPage)
  soup=BeautifulSoup(response.text,"html.parser")
  print(soup)
  schoolName=soup.find_all("h2", class_="search-result__title")
  for i in range (0,25):
    cur=schoolName[i].text
    schoolList.append(cur)
with open('nicheRanking.csv', 'w', newline='') as csvfile:
    fieldnames = ['ranking', 'college']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames) 
    writer.writeheader()

    for x in range(len(schoolList)):
        writer.writerow({'ranking': ranking, 'college': schoolList[x]})
        ranking+=1

```
