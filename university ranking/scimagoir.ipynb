```python
import csv
import requests
from bs4 import BeautifulSoup

lists=[]
country=[]
list2=[]
response = requests.get('https://www.scimagoir.com/rankings.php?sector=Higher+educ.&ranking=Overall&area=all')
soup=BeautifulSoup(response.text,"html.parser")
x=soup.find_all("td")

for ele in x: 
  lists.append(ele.text)
for count in range(2,len(lists)):
  if (count-2)%6==0:
    list2.append(lists[count])
  elif  (count-2)%6==1:
    country.append(lists[count])

for c in range (0,len(list2)):
  if list2[c].find("*")>0:
    list2[c]=list2[c].replace("*","")


with open('scimagoir.csv', 'w', newline='') as csvfile:
    fieldnames = ['name', 'country', 'rank']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for x in range(0,len(country)):
      writer.writerow({'name': list2[x], 'country': country[x], 'rank':x+1})



        
```
