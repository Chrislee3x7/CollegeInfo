```python
import csv
import requests
from bs4 import BeautifulSoup
ARWU=[]
QS=[]
THE =[]
ARTU=[]
name=[]
ranking=[]
country=[]
count=0
response = requests.get('http://research.unsw.edu.au/artu/')
soup=BeautifulSoup(response.text,"html.parser")
x=soup.find_all("td")

# for ele in x:
#   if count%8==1:
#     print(ele.text)
#   count+=1

for ele in x:
    ele=ele.text
    if count%8==1 :
        name.append(ele)
    if count%8==2:
        country.append(ele)
    if count%8==3:
        ARWU.append(ele)
    if count%8==4:
        QS.append(ele)
    if count%8==5:
        THE.append(ele)
    if count%8==7:
        ARTU.append(ele)
    count+=1
    if count>4000:
      break

with open('p9.csv', 'w', newline='') as csvfile:
    fieldnames = ['name', 'country','arwu','QS', 'THE','ARTU']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for x in range(400):
        writer.writerow({'name': name[x], 'country': country[x], 'arwu': ARWU[x],'QS':QS[x],'THE':THE[x],'ARTU':ARTU[x]})




```
